```python
from dotenv import load_dotenv
import os

load_dotenv()

db_user = os.getenv("DB_USER")
db_password = os.getenv("DB_PASSWORD")
db_host = os.getenv("DB_HOST")
db_port = os.getenv("DB_PORT")
db_name = os.getenv("DB_NAME")

#Connecting to Database using SQLalchemy 
from sqlalchemy import create_engine
import pandas as pd

engine = create_engine(f"postgresql+psycopg2://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}")
print("Connection successfuly")
```

```python
#Extracting Data from Fakestore api 
import requests
base_URL = "https://fakestoreapi.com/products"

#Fetching data 
def extracting_from_FakeStore_API(url):
    """This is a function to get data from api and return as dataframe"""

    response = requests.get(url)
    data = response.json()
    
    # Converting  to DataFrame
    df_products = pd.DataFrame(data)


    # Flatten 'rating' dict into separate columns
    df_products = pd.concat([df_products.drop(['rating'], axis=1),
                  df_products['rating'].apply(pd.Series).rename(columns=lambda x: f'rating_{x}')],
               axis=1)

    df_products .to_sql("products", engine, if_exists="replace", index=False)

    
    df_products.head()

    return df_products


```

```python
extracting_from_FakeStore_API(base_URL)
```

```python
#Loading data from api to the PostgreSQL database
def load(df_products):
     
  df_products.to_sql("products", engine, if_exists="replace", index=False)
  print("products loaded")


```

```python
df = extracting_from_FakeStore_API(base_URL)
load(df)
```
